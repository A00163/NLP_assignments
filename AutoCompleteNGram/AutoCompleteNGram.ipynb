```python
from hazm import sent_tokenize, word_tokenize, Normalizer, stopwords_list
punctuation_pattern = r'[^\w\s]'
all_comments = re.sub(punctuation_pattern, '', all_comments)
sentences = sent_tokenize(all_comments)
stopwords = stopwords_list()
normalizer = Normalizer()
word_list = []
cleaned_text = []
for sentence in sentences:
    # Tokenize the sentence into words
    words = word_tokenize(sentence)

    # Normalize the words
    normalized_words = [normalizer.normalize(word) for word in words]

    # Remove stopwords
    filtered_words = [word for word in normalized_words if word not in stopwords]

    # Store the processed words in the final list
    word_list.append(filtered_words)

    cleaned_text.append(" ".join(filtered_words))
```

```python
tokenized_sentences = []
for sentence in cleaned_text:
  tokenized_sentences.append(word_tokenize(sentence))
```

```python
word_counts = {}
for sentence in tokenized_sentences:
    for token in sentence:
        if token not in word_counts.keys():
            word_counts[token] = 1
        else:
            word_counts[token] += 1
```

```python
def get_words_frequency(tokenized_sentences, min_frequency=2):
  vocab = []
  for word, cont in word_counts.items():
     if cont >= min_frequency:
        vocab.append(word)
  return vocab
```

```python
def unk(tokenized_sentences, vocabulary, unknown_token="<unk>"):
    vocabulary = set(vocabulary)
    replaced_tokenized_sentences = []
    for sentence in tokenized_sentences:
        replaced_sentence = []
        for token in sentence:
            if token in vocabulary:
                replaced_sentence.append(token)
            else:
                replaced_sentence.append(unknown_token)
        replaced_tokenized_sentences.append(replaced_sentence)
    return replaced_tokenized_sentences
```

```python
def get_ngrams(tokens, n):
    start_tokens = ['<start>'] * (n - 1) if n > 1 else []
    end_token = ['</end>']
    full_tokens = start_tokens + tokens + end_token
    ngrams = [tuple(full_tokens[i:i + n]) for i in range(len(full_tokens) - n + 1)]
    return ngrams
```

```python
def bigram_estimate(word, bigram_counter):
  bigram_freq_dist = FreqDist(bigram_counter)
  bigram_good_turing = SimpleGoodTuringProbDist(bigram_freq_dist)
  probability = bigram_good_turing.prob(word)
  return probability

bigram_probabilities = {}
for bigram,count in bigram_counter.items():
    probability = bigram_estimate(bigram, bigram_counter)
    print(f"{bigram}: {probability:.5f}")
    bigram_probabilities[bigram] = probability
```

##Part3:A function to calculate the perplexity of each created model (trigram, bigram, unigram).

```python
def estimate_probability(word,n,n_gram_counts):
  if n==1:
    probability = unigram_estimate(word, n_gram_counts)
  elif n==2:
    probability = bigram_estimate(word, n_gram_counts)
  elif n==3:
    probability = trigram_estimate(word, n_gram_counts)
  return probability
```

```python
def calculate_perplexity(tokens,n,n_gram_count):
    N = len(tokens)
    product_pi = 1.0
    probability = 0
    for token in tokens:

        probability = estimate_probability(token,n,n_gram_count)

        if probability > 0:
            product_pi *= 1 / probability
        else:
            return float('inf')

    perplexity = product_pi**(1/float(N))

    return perplexity
```

## Part 4: Build an auto-complete system

```python
import random
def suggest_sentence(sentence, n_gram_models, n_gram_probabilities, num_words=12):
    random.seed(74)
    suggested_sentence = sentence.copy()
    n_gram_order = len(list(n_gram_models.keys())[0])
    suggested_words = set()

    while len(suggested_sentence) < num_words:
        if n_gram_order == 1:
            context = ()
        elif n_gram_order == 2:
            context = tuple(suggested_sentence[-1:])
        elif n_gram_order == 3:
            context = tuple(suggested_sentence[-2:])

        prob_dist = n_gram_probabilities
        possible_next_words = [ngram[-1] for ngram in prob_dist.keys() if ngram[:-1] == context]
        next_word_probs = [prob_dist[ngram] for ngram in prob_dist.keys() if ngram[:-1] == context]
        if next_word_probs:
            top_indices = sorted(range(len(next_word_probs)), key=lambda i: next_word_probs[i], reverse=True)[:7]

            next_word_index = random.choice(top_indices)
            next_word = possible_next_words[next_word_index]

            while next_word in suggested_words:
                top_indices.remove(next_word_index)
                if top_indices:
                    next_word_index = random.choice(top_indices)
                    next_word = possible_next_words[next_word_index]
                else:
                    break

            suggested_words.add(next_word)
        else:
            break

        suggested_sentence.append(next_word)

    return suggested_sentence
```
