```python
from hazm import Normalizer, sent_tokenize, word_tokenize, stopwords_list
import re
```

```python
file_path = '/content/hp_fa.txt'
with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
```

## Step 1: Remove extra blank spaces, parse text, and normalize

```python
normalizer = Normalizer()
normalized_text = normalizer.normalize(text)
```

```python
normalized_text[:10000]
```

##Replace ellipsis with a single period

```python
text = normalized_text.replace("...", ".")
```

```python
text[:1000]
```

#Step 2: Tokenize sentences into words


```python
from hazm import sent_tokenize, word_tokenize
text_sentences = sent_tokenize(text)
```

```python
print(text_sentences[:100])
```

```python
tokenized_words = [word_tokenize(sentence) for sentence in text_sentences]
```

```python
tokenized_words[:1000]
```

##Step 3: Remove punctuation marks


```python
def remove_punctuation(text):
    return re.sub(r'[^\w\sุุุยซยปโโโโโฆโโโูุก๏ทผูซูฌ]', '', text)
```

```python
tokenized_words_no_punctuation = [[remove_punctuation(word) for word in sentence] for sentence in tokenized_words]
```

##Step 4: Remove stop words

```python
persian_stopwords = set(stopwords_list())
tokenized_words_no_stopwords = [[word for word in sentence if word not in persian_stopwords] for sentence in tokenized_words_no_punctuation]
```

##Step 5: Remove emojis

```python
def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emoticons
                               u"\U0001F300-\U0001F5FF"  # Symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # Transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # Flags (iOS)
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
tokenized_words_no_emojis = [[remove_emojis(word) for word in sentence] for sentence in tokenized_words_no_stopwords]
```

```python
tokenized_words_no_emojis[:100]
```

##Step 6: Apply lemmatization





```python
from hazm import Lemmatizer
lemmatizer = Lemmatizer()
```

```python
# lemmatized_text = [lemmatizer.lemmatize(word) for token in tokenized_words_no_emojis]
lemmatized_words = [[lemmatizer.lemmatize(word) for word in sentence] for sentence in tokenized_words_no_emojis]
```

```python
for sentence in lemmatized_words:
    print(' '.join(sentence))
```

```python
output_file_path = "/content/output.txt"
with open(output_file_path, "w", encoding="utf-8") as file:
    for sentence in tokenized_words_no_emojis:
        file.write(' '.join(sentence) + '\n')
```
